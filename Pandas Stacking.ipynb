## Stack Example
**Stack** turns the specified row into a column

```python
import pandas as pd
import os
os.getcwd()
df = pd.read_excel(r'c:\users\rmainer\desktop\stocks.xlsx', header=[0,1])  # specifying that header is rows 0,1
#df = pd.read_excel("stocks.xlsx", header=[0,1])
df
```

```python
df_stacked = df.stack(level=0)  # here we turned the first header (level0) into a column
df_stacked
```

```python
df_stacked.unstack()  # it's possible to immediatly unstack
```
